```python
X=df.iloc[:,0:1].values
y=df.iloc[:,1].values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X,y)
```

```python
lr.predict([[3]])
```

```python
y_p=lr.predict(X)
y_p
```

```python
y
```

```python
plt.scatter(X,y,color='red',label="Actual Values")
plt.plot(X,y_p,color='k',label="Predicted Values")
plt.xlabel("Years of Experience")
plt.ylabel("Expected Salary")
plt.title("Linear Regression to predict salary based on years of experience")
plt.legend()
plt.show()
```

```python
#Regression Metrics
#MAE : Mean Absolute Error
#MSE : Mean Squared Error
#RMSE : Root Mean Squared Error
#R2 score : Co-efficient of Determination
#Adjusted R2 score
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_score(y,y_p)
```
